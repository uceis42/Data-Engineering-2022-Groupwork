```python

import pandas as pd
import numpy as np
```

```python
df2 = pd.read_csv('NameList300.csv',encoding='latin1')
df1 = pd.read_csv('df_detail_v3.csv',encoding='latin1')
```

```python
df3 = pd.concat([df1, df2], axis=1)
```

```python
df3['name_letter'] = df3['Name'].str.replace(" ","").str.replace("'","").str.lower()
```

```python
df3['name_letter']
```

```python
df3 = df3.drop(columns = ['Unnamed: 0.2','Unnamed: 0.1','Unnamed: 0'])
```

```python
df3
```

```python
df3_new = df3[df3['rating'].notna()]
df3_new
```

```python
df3_new_withoutprice = df3_new[~df3_new['price_range'].str.startswith("?",na=False)]
df3_new_withoutprice
```

```python
df3_new_withoutprice['features'] = df3_new_withoutprice['cuisines']
df3_new_withoutprice['cuisines'] = df3_new_withoutprice['price_range']
df3_new_withoutprice['price_range'] = np.nan
```

```python
df3_new = df3_new.drop(df3_new[~df3_new['price_range'].str.startswith("?",na=False)].index)
```

```python
df3_new
```

```python


df4 = pd.concat([df3_new, df3_new_withoutprice])

df4
```

```python
df4 = df4.sort_index()
df4['n_reviews'] = df4['n_reviews'].str[:-7].str.replace(',','')
df4['about'] = df4['about'].str[:200]
df4['features'] = df4['features'].str[:200]
df4
```

```python
df4.to_csv('df_detail_final.csv')
```
