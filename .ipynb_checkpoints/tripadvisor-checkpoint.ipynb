```python

def get_name(soup):  
    #a = soup.find(class_="bHGqj Cj b")
    n_list =[]
    level1 = soup.find_all("div", {"class": "list",'id':'EATERY_OVERVIEW_BOX'})
      
    for page_num in range(1,31):
        num_str = str(page_num)+ '_list_item'
        for item  in level1:   
            item=soup.find('div', attrs={'data-test': num_str})
            text = item.find('a',{'class':'bHGqj Cj b'})
            realtext = text.find_all(text=True)[4]
            n_list.append(realtext)

 
    #k = str(a.string).replace('	', '').replace('\n', '').replace('\r', '')
    return n_list


name_list = get_name(soup)
df = pd.DataFrame(name_list,columns=['Name'])
```

```python
def get_url(soup):  
    #a = soup.find(class_="bHGqj Cj b")
    U_list =[]
    level1 = soup.find_all("div", {"class": "list",'id':'EATERY_OVERVIEW_BOX'})
      
    for page_num in range(1,31):
        num_str = str(page_num)+ '_list_item'
        for item  in level1:   
            item=soup.find('div', attrs={'data-test': num_str})
            text = item.find('a',{'class':'bHGqj Cj b'})
            URL= text.get('href')
            U_list.append(URL)
    return U_list


#URL_list = description(soup)

url_list = get_url(soup)
url_full_list= []
for item in url_list:
    item = 'https://www.tripadvisor.com'+str(item)
    url_full_list.append(item)
df["url"] = url_full_list
```
